```python
# PRIMER FILTRO POR CVE_CLPV
filtroUNO =df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtroUNO
```

```python
filtroUNO.to_csv("FiltroUNO.csv")
# En está sección descargo el csv
from google.colab import files   
files.download("FiltroUNO.csv")
```

```python
# SEGUNDO FILTRO POR CVE_VEND
filtroDOS =df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtroDOS 
```

```python
filtroDOS.to_csv("FiltroDOS.csv")
# En está sección descargo el csv
from google.colab import files   
files.download("FiltroDOS.csv")
```

```python
# TERCER FILTRO POR FECHAS DE "28/02/2022”
filtroTRES = df[df["FECHA_ENT"] == "28/02/2022 23:59:59.000"]
filtroTRES
```

```python
filtroTRES.to_csv("FiltroTRES.csv")
# En está sección descargo el csv
from google.colab import files   
files.download("FiltroTRES.csv")
```
